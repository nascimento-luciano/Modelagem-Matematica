```python
# d) Desenvolva um programa que, aplicando o método da bisecção, resolva as seguintes questões:

def bissecao(f,a,b,tol=1e-9,Nmax=50):
  n=1
  while n<Nmax:
    c=(b+a)/2
    #print(n,'f(c)=', f(c))
    if f(c)==0 or ((b + a)/2)<tol:
       print('raiz=',c)
       print('erro=',f(c))
       if f(c)<=1e-2:
         return c
    elif sign(f(c)*f(a))>0:
      a=c
    else:
      b=c
    n +=1
print('Raízes não encontradas')
```
